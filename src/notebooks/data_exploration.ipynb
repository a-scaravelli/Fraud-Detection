```python
def is_unique(s):
    a = s.to_numpy() 
    return (a[0] == a).all()


valore_nullo = ' '
nulle = {}
da_droppare = []
for column in df.columns:
  #check nulli
  #nulli += df[df[column] == valore_nullo].shape[0]
  if df[df[column] == valore_nullo].shape[0] != 0:
    nulle[column] = df[df[column] == valore_nullo].shape[0]

  #check monovalori
  test = is_unique(df[column])
  if test:
    #print(column)
    da_droppare.append(column)
  
print(nulle)
print(da_droppare)
```

```python
df.drop(da_droppare,axis = 1)
```

```python
corr = df.corr()


components = list()
visited = set()
print(df.columns)
for col in df.columns:
    if col in visited:
        continue

    component = set([col, ])
    just_visited = [col, ]
    visited.add(col)
    while just_visited:
        c = just_visited.pop(0)
        for idx, val in corr[c].items():
            if abs(val) > 0.80 and abs(val) != 1.000 and idx not in visited:
                just_visited.append(idx)
                visited.add(idx)
                component.add(idx)
    components.append(component)

for component in components:
    if len(component) > 1:
      plt.figure(figsize=(12,8))
      sns.heatmap(corr.loc[component, component], cmap="Reds")
```
